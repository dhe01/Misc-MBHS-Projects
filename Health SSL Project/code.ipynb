```python
locationlist = []
for key in dist.keys():
    key1 = key.split(",")
    locationlist.append(np.asarray([int(key1[0]), int(key1[1])]))

locationlist
```

```python
def probfunx(d):
    return(1/(1+np.exp(d))*2)
```

```python
df = pd.DataFrame(columns=["Location", "Value"])

for i in locationlist:
    currsum = 0
    for j in locationlist:
        currsum += probfunx(np.linalg.norm(i-j)) * dist[str(j[0]) + "," + str(j[1])]
    df.loc[len(df)] = [i, currsum]

#print(maxloc, max2loc, max3loc, max4loc, max5loc)
df = df.sort_values("Value", ascending=False).reset_index(drop=True)
df.to_excel("optimallocations.xlsx")
```

```python
df2 = pd.DataFrame(columns=["Location 1", "Location 2", "Value"])

for i in range(len(locationlist)):
    maxsum = 0
    maxj = 0
    for j in range(i,len(locationlist)):
        currsum = 0
        for k in locationlist:   
            currsum += probfunx(min(np.linalg.norm(locationlist[i]-k), np.linalg.norm(locationlist[j]-k))) * dist[str(k[0]) + "," + str(k[1])]
        if currsum > maxsum:
            maxsum = currsum
            maxj = j
    df2.loc[len(df2)] = [locationlist[i], locationlist[maxj], maxsum]
    print(locationlist[i])
```

```python
df2 = df2.sort_values("Value", ascending=False).reset_index(drop=True)
df2.to_excel("optimallocations2.xlsx")
df2
```
