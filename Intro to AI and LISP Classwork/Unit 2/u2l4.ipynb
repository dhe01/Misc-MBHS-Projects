```python
house = pd.read_csv("Loudoun_Housing_data.csv")

realx = house[["Bedrooms", "Bathrooms", "Square Feet"]]
x = np.c_[np.ones(len(realx)), realx]

y = house["Price"]

w = [0, 60000, 50000, 500]

lr = 0.00000013

costx = []
costy = []

flag = True

c = 0

while flag:
    c += 1
    pred = np.dot(x, w)
    diff = pred - y
    grad = np.dot(diff, x)/len(x)
    cost = np.sum(diff**2)
    
    w = w - lr * grad
    costx.append(c)
    costy.append(cost)
        
    if c > 10000:
        break
```

```python
fig = make_subplots(rows = 1, cols = 1)
fig.add_trace(go.Scatter(x = costx, y = costy, mode = "markers+lines", name = "Real"), row = 1, col = 1)
fig.show()
```

```python
clf = linear_model.LinearRegression()
clf.fit(np.asarray(house[["Bedrooms", "Bathrooms", "Square Feet"]]), [[i] for i in house["Price"]])

coefficients = list(clf.coef_)
coefficients
```

```python
print(w)
print(np.dot(x, w))
print(house["Price"])
print(np.dot(x, w) - house["Price"])
```

```python
house2 = pd.read_csv("Loudoun_Housing_data.csv")
house2["Bedrooms"] = normalize(house2["Bedrooms"])
house2["Bathrooms"] = normalize(house2["Bathrooms"])
house2["Square Feet"] = normalize(house2["Square Feet"])
house2["Price"] = normalize(house2["Price"])

realx = house2[["Bedrooms", "Bathrooms", "Square Feet"]]
x = np.c_[np.ones(len(realx)), realx]

y = house2["Price"]

w = [0, 60000, 50000, 300000]

lr = 0.9

costx = []
costy = []

flag = True

c = 0

while flag:
    c += 1
    pred = np.dot(x, w)
    diff = pred - y
    grad = np.dot(diff, x)/len(x)
    cost = np.sum(diff**2)
    
    w = w - lr * grad
    costx.append(c)
    costy.append(cost)
        
    if c > 10000:
        break
```

```python
fig = make_subplots(rows = 1, cols = 1)
fig.add_trace(go.Scatter(x = costx, y = costy, mode = "markers+lines", name = "Real"), row = 1, col = 1)
fig.show()
```

```python
clf = linear_model.LinearRegression()
clf.fit(np.asarray(house2[["Bedrooms", "Bathrooms", "Square Feet"]]), [[i] for i in house2["Price"]])

coefficients = list(clf.coef_)
coefficients
```
