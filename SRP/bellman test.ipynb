```python
def lamb(n,a):
    return((1.1-0.1*n)*(9-0.25*a))

def p(n, s, a):
    problist = []
    for j in range(s+1):
        if j == 0:
            j = s
            prob = 1
            for i in range(1, s):
                prob -= np.e ** (-lamb(n,a)) * (lamb(n,a)) ** (s-i)/(np.math.factorial(s-i))
        else:
            prob = np.e ** (-lamb(n,a)) * (lamb(n,a)) ** (s-j)/(np.math.factorial(s-j))
        problist.append(prob)
    return(problist)

def u(n, s):
    if n == N:
        return([r(n, s, scrap, 0), "done"])
    else:
        for s in state:
            optval = 0
            optact = 0
            for a in action:
                val = 0
                prob = p(n,s,a)
                for j in range(s+1):
                    val += prob[j] * (j + int(u(n+1, j)[0]))
                    if val > optval:
                        optval = val
                        optact = j
    print(n,s)
    return([optval, optact])

def r(n, s, a, j):
    if n == N:
        return(s * a)
    else:
        return(a * (s-j) - hold * j)
```

```python
N = 6
start = 15

scrap = 0
hold = 2
action = [0, 20, 23, 25, 27, 30, 35]
state = [i for i in range(16)]

u(0, 15)
```

```python
lam = lambda a, n: (1.1-0.1*n)*(9-0.25*a)
poisson = lambda a,n,s,i: (np.exp(-lam(a,n))) * ((lam(a,n)) ** (s-i))/(np.math.factorial(s-i))

def prob(n, s, a):
    
    problist = []
    
    # Iterates through stock remaining
    for j in range(s+1):
        
        # If there is no stock remaining, the demand was >= s
        if j == 0:
            prob = 1
            
            # Complementary counting
            for i in range(1, s+1):
                prob -= poisson(a,n,s,i)
        
        else:
            # Cost is distributed in poisson manner
            prob = poisson(a,n,s,j)
        
        problist.append(prob)
        
    return(problist)

def reward(n, s, a, j):
    
    # If no more epochs left
    if n == N:
        # Sell remaining inventory for scrap value
        return(s * scrap)
    
    # If more time
    else:
        # Money from sales - costs for storage
        return(a * (s-j) - hold * j)
    
def u(n, oldval, allact):
    
    # At the end of time, start of program
    if n == N:
        # Returns a list of u(n,s) for all possible states s
        return(u(n-1, [reward(n, s, 0, 0) for s in range(len(state)+1)], []))
    
    # At the start of time, end of program
    elif n == 0:
        return(allact)
    
    else:
        
        # Declare lists that contain optimal moves
        currval = []
        curract = []
        
        for s in state:
            optval = 0
            optact = 0
            
            for a in action:
                val = 0
                problist = prob(n,s,a)
                                
                for j in range(s+1):
                    
                    val += problist[j] * (reward(n, s, a, j) + oldval[j])
                    
                if val > optval:
                    optval = val
                    optact = a
                
            currval.append(optval)
            curract.append(optact)
        
        allact.append(curract)
        return(u(n-1, currval, allact))           
```
