```python
enroll = pd.read_excel("Vermont Data/Vermont Enrollment Info.xlsx")
enroll = enroll[enroll["SchoolYear"] == 2022][["SchoolIdentifier", "OrganizationName", "Preschool", "Preschool", "KindergartenFullTime", "KindergartenPartTime", "FirstGrade", "SecondGrade", "ThirdGrade", "FourthGrade", "FifthGrade", "SixthGrade", "SeventhGrade", "EighthGrade", "NinethGrade", "TenthGrade", "EleventhGrade", "TwelvethGrade", "Total"]].reset_index(drop=True)

info = pd.read_csv("Vermont Data/Vermont Science Data.csv")
info = info[info["SY"] == 2022][info["AssessGroup"] == "All Students"][info["Indicator_Label"] == "Average Scaled Score"].reset_index(drop=True)

sciscore = {}
infoorg = list(info["OrganizationIdentifier"])

for school in enroll["SchoolIdentifier"]:
    if school in infoorg:
        scilist = (info[info["OrganizationIdentifier"] == school]["SchoolValue"])
        totalsum = 0
        totalcount = 0
        for val in scilist:
            if "." in val:
                totalsum += float(val)
                totalcount += 1
        
        if totalsum == 0:
            sciscore[school] = 0
        else:
            sciscore[school] = totalsum/totalcount

    else:
        sciscore[school] = 0
sciscore
```

```python
enroll["Scores"] = sciscore.values()
enroll = enroll[enroll["Scores"] != 0]
enroll
```
